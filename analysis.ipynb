# ArcGIS Learn Autoresearch Analysis

Review validation mAP, precision, and recall trends from `results.tsv`.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('results.tsv', sep='\t')
for column in ['val_map', 'val_precision', 'val_recall', 'training_minutes']:
    df[column] = pd.to_numeric(df[column], errors='coerce')

df
```

```python
print('Run count:', len(df))
print('\nStatus counts:')
print(df['status'].fillna('<missing>').value_counts().to_string())

if not df['val_map'].dropna().empty:
    best_idx = df['val_map'].idxmax()
    best_row = df.loc[best_idx]
    print('\nBest validation run:')
    print(best_row[['run_id', 'proposal_title', 'primary_change', 'architecture', 'backbone', 'chip_size', 'val_map', 'val_precision', 'val_recall', 'status']].to_string())
```

```python
plot_df = df.copy()
plot_df['run_number'] = range(1, len(plot_df) + 1)

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(plot_df['run_number'], plot_df['val_map'], marker='o', label='Validation mAP')
ax.set_title('Validation mAP Over Time')
ax.set_xlabel('Run Number')
ax.set_ylabel('mAP')
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()
```

```python
metric_df = df[['proposal_title', 'primary_change', 'architecture', 'backbone', 'chip_size', 'val_map', 'val_precision', 'val_recall', 'training_minutes', 'status', 'notes']].copy()
metric_df.sort_values(['status', 'val_map'], ascending=[True, False])
```
